```python
import yfinance as yf
import pandas as pd
```

```python
sp_ticker = yf.Ticker("^GSPC")
sp_data = sp_ticker.history(period = "1y")
print(sp_data.head())
```

```python
target_date = pd.to_datetime('2024-12-05')
# sp_data uses a DatetimeIndex (no 'Date' column). Compare against the index instead.
# Use .date() to avoid timezone-aware vs naive Timestamp mismatches.
rows_for_specific_date = sp_data[sp_data.index.date == target_date.date()]
print(rows_for_specific_date)
```
