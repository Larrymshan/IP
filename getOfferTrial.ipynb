```python
import pandas as pd
import requests
```

```python
def fetch_data():
    url = "https://api.nasdaq.com/api/ipo/calendar"
    headers = {
        "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/118.0.0.0 Safari/537.36",
        "Accept-Language": "en-US,en;q=0.9",
        "Origin": "https://www.nasdaq.com",
        "Referer": "https://www.nasdaq.com/"
    }
    
    try:
        response = requests.get(url, headers=headers, timeout=10)
        response.raise_for_status()
        data = response.json()
        return data['data']['priced']['rows']
    except requests.RequestException:
        print("Request failed, either blocked or bad")
        return []
```

```python
ipo_list = fetch_data()

if ipo_list:
    df = pd.DataFrame(ipo_list)
    clean_df = df[['proposedTickerSymbol', 'companyName', 'proposedSharePrice', 'pricedDate', 'proposedExchange']]
    print(clean_df.head())
    print(df)
else:
    print("No data available.")
```
