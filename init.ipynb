```python
import numpy as np
import pandas as pd
```

```python
import os

file_path = os.path.join('data', 'IPODataFull.csv')

df = pd.read_csv(file_path, encoding="ISO-8859-1", low_memory=False)

df.head(10)
```

```python
print(df.shape)

df_deduped = df.drop_duplicates(subset="Symbol")

print(df_deduped.shape)
```

```python
#Ticker, IPO Date, Offer Price, First Day Close

df['date'] = pd.to_datetime(df_deduped[['Year', 'Month', 'Day']])

df2 = pd.DataFrame()

df2['Ticker'] = df_deduped['Symbol']
df2['IPO Date'] = df['date']
df2[['Offer Price', 'First Day Close', 'Day2', 'Day3', 'Day4', 'Day5', 'Day6', 'Day7', 'Day8', 'Day9', 'Day10']] = df_deduped[['openDay0', 'closeDay0', 'closeDay1', 'closeDay2', 'closeDay3', 'closeDay4', 'closeDay5', 'closeDay6', 'closeDay7', 'closeDay8', 'closeDay9']]
df2 = df2.sort_values(by='IPO Date')

df2.head(20)
```
